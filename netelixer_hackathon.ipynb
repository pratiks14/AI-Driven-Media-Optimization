```python
def identify_outliers(df, column):

    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers
```

**Preprocess Landing Data**
1. Remove duplicate rows
2. For a given user delete rows after the first conversion( Such rows will ignored for the construction of transition matrix for markov chain)
3. Delete the user data whose touchpoints dont include any of three ad platform

```python
#Delete users who dont have touchpoints with the following sources
valid_sources = ['bing' ,'google', 'facebook']
sorted_land_df['Valid Source'] = sorted_land_df['Source'].isin(valid_sources)

sorted_land_df.loc[
    sorted_land_df['Valid Source'] &
    sorted_land_df['Source'].isin(['google', 'bing']) &
    sorted_land_df['Campaign Type'].isna(),
    'Valid Source'
] = False
valid_source_users = sorted_land_df.groupby('User Id')['Valid Source'].sum()
user_id_without_valid_source = valid_source_users[valid_source_users == 0].index
sorted_land_df = sorted_land_df[~sorted_land_df['User Id'].isin(user_id_without_valid_source)]
```

```python
#Delete Duplicate rows
temp_df = sorted_land_df[['User Id', 'Website Landing Time','Source' , 'Campaign Type']]
unique_indices = temp_df[~temp_df.duplicated()].index
sorted_land_df = sorted_land_df.loc[unique_indices].reset_index(drop=True)
```

# **Markov chain Attribution Model**

# Construction Of Weighted Transition Matrix
1. We need  Conversion to clicks ratio. This will act as weights as we add record of channel transition for each user.
2. Add Co

```python
google_df['CTR'] = google_df['Clicks']/google_df['Impressions']
google_df['CR'] = google_df['Conversions']/google_df['Clicks']
google_df['RC'] = google_df['Revenue']/google_df['Cost']

meta_df['CTR'] = meta_df['Clicks']/meta_df['Impressions']
meta_df['CR'] = meta_df['Conversions']/meta_df['Clicks']
meta_df['RC'] = meta_df['Revenue']/meta_df['Cost']

micro_df['CTR'] = micro_df['Clicks']/micro_df['Impressions']
micro_df['CR'] = micro_df['Conversions']/micro_df['Clicks']
micro_df['RC'] = micro_df['Revenue']/micro_df['Cost']
```

```python
google_df.loc[google_df['Impressions']== 0,'CTR'] = 0
google_df.loc[google_df['Clicks']== 0,'CR'] = 0
google_df.loc[google_df['Cost']== 0,'RC'] = 0

meta_df.loc[meta_df['Impressions']== 0,'CTR'] = 0
meta_df.loc[meta_df['Clicks']== 0,'CR'] = 0
meta_df.loc[meta_df['Cost']== 0,'RC'] = 0

micro_df.loc[micro_df['Impressions']== 0,'CTR'] = 0
micro_df.loc[micro_df['Clicks']== 0,'CR'] = 0
micro_df.loc[micro_df['Cost']== 0,'RC'] = 0
```

**Transition Matrix**

```python
from collections import defaultdict

# Initialize transition counts, including campaign types
transition_counts = defaultdict(lambda: defaultdict(int))
cust_cnt = 0

# Loop through customer journeys, including channel + campaign and conversion
for customer in sorted_land_df['User Id'].unique():
    cust_cnt += 1
    if cust_cnt % 1000 == 0:
      break
    customer_journey = sorted_land_df[sorted_land_df['User Id'] == customer]

    for i in range(len(customer_journey)):
        next_channel = ""
        if customer_journey.iloc[i]['Is Converted']:
          next_channel = 'Conversion'
        else:
          if i + 1 < len(customer_journey):
            if customer_journey.iloc[i + 1]['Valid Source']:
              next_channel = customer_journey.iloc[i + 1]['Source']
            else:
              next_channel = customer_journey.iloc[i + 1]['Channel']
          else:
            next_channel = 'No Conversion'
        if customer_journey.iloc[i]['Valid Source']:
            channel = customer_journey.iloc[i]['Source']
            if channel in ['google','bing']:
              campaign_type = customer_journey.iloc[i]['Campaign Type']
              print("Channel. : " , channel)
              if channel == 'google':

                print("Campaign",campaign_type)
                print("Date " ,customer_journey.iloc[i]['Date'])
                google_cr = google_df[(google_df['Campaign type'] == campaign_type) & (google_df['Date'].dt.date == customer_journey.iloc[i]['Date'])]['CR'].values[0]
                transition_counts[channel][next_channel] += google_cr
              else:
                micro_cr = micro_df[(micro_df['Campaign type'] == campaign_type) & (micro_df['Date'].dt.date == customer_journey.iloc[i]['Date'])]['CR'].values[0]
                transition_counts[channel][next_channel] += micro_cr
            else:
              meta_cr = meta_df[(meta_df['Date'].dt.date == customer_journey.iloc[i]['Date'])]['CR'].values[0]
              transition_counts[channel][next_channel] += meta_cr

        else:
            channel = customer_journey.iloc[i]['Channel']
            transition_counts[channel][next_channel] += 1

        if next_channel == 'Conversion':
          break

    # break

# Convert counts to probabilities based on CTR or conversion rate
transition_matrix = {}
for channel_from, transitions in transition_counts.items():
    total_transitions = sum(transitions.values())
    transition_matrix[channel_from] = {}
    for channel_to, count in transitions.items():
        transition_matrix[channel_from][channel_to] = (count / total_transitions)

# Convert transition matrix to a pandas DataFrame for visualization
transition_matrix_df = pd.DataFrame(transition_matrix)
print(transition_matrix_df)
```
